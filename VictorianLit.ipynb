# VictorianLit

Victorian Lit Dataset for Machine Learning-Based Sentiment Analysis of Victorian Literary Texts | by Hoyeol Kim

```
import torch
import numpy as np
import pandas as pd

from tqdm import tqdm

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```
df=pd.read_csv('https://github.com/elibooklover/VictorianLit/raw/master/VictorianLit.csv')
df.head()
```

```
df.describe()
```

```
sns.set(style = "darkgrid" , font_scale = 1.2)
sns.countplot(df.label)
```

```
fig,(ax1,ax2,ax3,ax4,ax5)=plt.subplots(1,5,figsize=(20,8))
text_len=df[df['label']==0]['sentences'].str.len()
ax1.hist(text_len,color='red')
ax1.set_title('Very Negative')
text_len=df[df['label']==1]['sentences'].str.len()
ax2.hist(text_len,color='orange')
ax2.set_title('Negative')
text_len=df[df['label']==2]['sentences'].str.len()
ax3.hist(text_len,color='green')
ax3.set_title('Neutral')
text_len=df[df['label']==3]['sentences'].str.len()
ax4.hist(text_len,color='skyblue')
ax4.set_title('Positive')
text_len=df[df['label']==4]['sentences'].str.len()
ax5.hist(text_len,color='blue')
ax5.set_title('Very Positive')

fig.suptitle('Characters in the VictorianLit Dataset')
plt.show()
```

```
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(20,10))
word=df[df['label']==(0 or 1)]['sentences'].str.split().apply(lambda x : [len(i) for i in x])
sns.distplot(word.map(lambda x: np.mean(x)),ax=ax1,color='red')
ax1.set_title('Texts with Negative Sentiment')
word=df[df['label']==(3 or 4)]['sentences'].str.split().apply(lambda x : [len(i) for i in x])
sns.distplot(word.map(lambda x: np.mean(x)),ax=ax2,color='green')
ax2.set_title('Texts with Positive Sentiment')
fig.suptitle('Average Word Length')
```
